> ## All the maps created are stored in the folder *viz_created*

```python
import pandas as pd
import numpy as np

import folium
print(folium.__version__)
```

```python
# Import data
data = pd.read_csv('data_created/data_ready_for_viz.csv', names=['id','District', 'Canton', 'Year', 'Month', 'Day', 'Score', 'ScoreBool'])
data.index = data.id
data.head()
```

# Canton

```python
cn = data.groupby('Canton').mean().ScoreBool
cn
```

```python
canton_geo = r'data/ch-cantons.topojson.json'

# Create the map and focus on Switzerland 
canton_map = folium.Map(location=[46.8, 8.28], zoom_start=8)


# Plot the map with colormap yellow orange red and manually set the scale. 
canton_map.choropleth(geo_path = canton_geo, 
                      data = cn,
                      columns = ['Canton', 'sentiment'],
                      key_on = 'feature.id',
                      topojson = 'objects.cantons',
                      threshold_scale = [-0.4, -0.2, 0, 0.2, 0.4],
                      fill_color = 'RdYlGn', 
                      fill_opacity = 0.7, 
                      line_opacity = 0.2,
                      legend_name = 'sentiment'
                 )

path = "viz_created/cantons.html"
canton_map.create_map(path)
    
canton_map
```

> **Rstigrben meine Freunden !**

> We can notice a very clear separtion between the french and german parts of Switzerland. Jura is the most *negative* one and Glarus the most *positive* one. The Basel-Stadt canton is the only german-speaking canton that has a negative score (between -0.2 and 0)

# District

```python
ds = data.groupby('District').mean().ScoreBool
ds
```

```python
sentiment_default = 0

# Load all district
df_districts = pd.read_table('data/ch_districts_capital.txt', sep=',')
district_list = df_districts.id_ofs.values
len(district_list)

# If District is missing
for x in district_list:
    if x not in list(ds.keys()):
        ds = ds.append(pd.Series([sentiment_default], index=[x]))
```

```python
# Districts geofile
district_geo = r'data/ch-districts.topojson.json'

# Create the map and focus on Switzerland 
district_map = folium.Map(location=[46.8, 8.28], zoom_start=8)

# Plot the map with colormap yellow orange red and manually set the scale. 
district_map.choropleth(geo_path = district_geo,
                      data = ds,
                      columns = ['District', 'sentiment'],
                      key_on = 'feature.id',
                      topojson = 'objects.districts',
                      threshold_scale = [-0.4, -0.2, 0, 0.2, 0.4],
                      fill_color = 'RdYlGn', 
                      fill_opacity = 0.7, 
                      line_opacity = 0.2,
                      legend_name = 'sentiment')

path = "viz_created/districts.html"
district_map.create_map(path)

district_map
```

> The district division of tweets assert the findings on the Canton viz. The *Rstigrben* is still present, and we can also notice that in the Valais and Fribourg cantons, the only "positive" district are the *german* ones !

> We also notice that there isn't clear differences between urban and rural districts. 

# Year

```python
for y in data.Year.unique():
    data_y = data[data.Year == y]
    print(y, len(data_y))
```

```python
years = [2012,2013,2014,2015,2016]
y_maps = list()
for y in years:
    data_y = data[data.Year == y]
    ds = data_y.groupby('District').mean().ScoreBool
    
    sentiment_default = -1

    # If District is missing
    for x in district_list:
        if x not in list(ds.keys()):
            ds = ds.append(pd.Series([sentiment_default], index=[x]))
            
    # Create the map and focus on Switzerland 
    district_map = folium.Map(location=[46.8, 8.28], zoom_start=8)

    # Plot the map with colormap yellow orange red and manually set the scale. 
    district_map.choropleth(geo_path = district_geo,
                          data = ds,
                          columns = ['District', 'sentiment'],
                          key_on = 'feature.id',
                          topojson = 'objects.districts',
                          threshold_scale = [-0.4, -0.2, 0, 0.2, 0.4],
                          fill_color = 'RdYlGn', 
                          fill_opacity = 0.7, 
                          line_opacity = 0.2,
                          legend_name = 'sentiment')
    
    path = "viz_created/year_"+str(y)+".html"
    district_map.create_map(path)
    y_maps.append(district_map)
```

> For the five years we have build a map, the rostigraben is always present.There is no *evolution* through the years, but 2013 was apparently a bad year, with a lot of negative tweets (almost all districts in Romandie).
